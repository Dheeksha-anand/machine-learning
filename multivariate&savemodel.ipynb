```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df=pd.read_csv("/content/drive/MyDrive/dataset/home.csv")
df
```

```python
import math
median_bedrooms = math.floor(df.bedrooms.median())
median_bedrooms
```

```python
df.bedrooms= df.bedrooms.fillna(median_bedrooms)
df
```

```python
model=linear_model.LinearRegression()
model.fit(df[['area','bedrooms','age']],df.price)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[3000,3,40]])
```

```python
#methods for save the file
#pickle metho
import pickle
```

```python
with open('model_pickle','wb') as f:
  pickle.dump(model,f)
```

```python
with open('model_pickle','rb') as f:
  mp=pickle.load(f)
```

```python
mp.predict([[3000,3,40]])
```

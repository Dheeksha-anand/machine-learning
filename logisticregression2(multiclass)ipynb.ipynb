```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
dir(iris)
```

```python
x=iris.data
y=iris.target
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
new_samples = [[5.1, 3.5, 1.4, 0.2],
               [9.0, 8.2, 7.0, 9.0],
               [5.1, 3.5, 1.4, 0.2]]

# Standardize the new samples
new_samples = scaler.transform(new_samples)

# Predict the category of the new samples
predictions = model.predict(new_samples)

# Map the predictions to species names
predicted_species = [iris.target_names[p] for p in predictions]
print("Predicted species:", predicted_species)
```

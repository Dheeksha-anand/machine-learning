```python
import pandas as pd
import numpy as np
df=pd.read_csv("/content/drive/MyDrive/dataset/titanic.csv")
df
```

```python
#checking the missing values
df.isna().sum()
```

```python
df["Age"].mean()
```

```python
#Filling the missing values
df["Age"].fillna(np.mean(df["Age"]),inplace=True)
```

```python
df.isna().sum()
```

```python
print(df.columns)
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
```

```python
df
```

```python
inputs=df.drop('Survived',axis='columns')
target=df.Survived
```

```python
from sklearn.preprocessing import LabelEncoder
le_sex=LabelEncoder()
```

```python
inputs['Sex']=le_sex=le_sex.fit_transform(inputs['Sex'])
inputs
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(inputs,target,test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
output=model.predict([[3,1,35.000000,8.0500]])
```

```python
if output == 1:
    print(f"The model predicts that the person has a high possibility to survie.")
else:
    print(f"The model predicts that the person has a low possibility to survie")
```
